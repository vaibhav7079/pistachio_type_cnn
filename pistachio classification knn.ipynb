```python
from sklearn.pipeline import Pipeline , make_pipeline
from sklearn.compose import ColumnTransformer , make_column_transformer
from sklearn.preprocessing import RobustScaler , StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest , chi2 , f_classif
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = df.drop('Class',axis=1)
```

```python
df['Class'].unique()
```

```python
y = df['Class'].map({'Kirmizi_Pistachio':0,'Siirt_Pistachio':1})
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
sk = SelectKBest(f_classif, k=20)
rs = RobustScaler()
ss = StandardScaler()
```

```python
col1=ColumnTransformer(transformers=[('Outlier_Treatment',rs,X.columns),
                                    ('Scaler',ss,X.columns)])
```

```python
p1 = Pipeline(steps=[('Column_transaformation',col1),
                    ('SelectKbest',sk),
                    ('Linear_Regression',LogisticRegression())])
```

```python
p1.fit(X_train,y_train)
```

```python
train_preds = p1.predict(X_train)
test_preds = p1.predict(X_test)
```

```python
from sklearn.metrics import classification_report , confusion_matrix
```

```python
print("Train Report ==>")
print(classification_report(y_train,train_preds))
```

```python
print("Test Report ==>")
print(classification_report(y_test,test_preds))
```

```python
sns.heatmap(data = confusion_matrix(y_test,test_preds),annot=True)
plt.title("Test Confusion Matrix",c='r',fontsize=15)
plt.show()
```

```python
sns.heatmap(data = confusion_matrix(y_train,train_preds),annot=True)
plt.title("Train Confusion Matrix",c='r',fontsize=15)
plt.show()
```

```python
col2 = make_column_transformer((RobustScaler(),X.columns),
                               (StandardScaler(),X.columns))
```

```python
p2 = make_pipeline(col2,KNeighborsClassifier())
```

```python
p2.fit(X_train,y_train)
```

```python
train_preds = p2.predict(X_train)
test_preds = p2.predict(X_test)
```

```python
print("Train Report ==>")
print(classification_report(y_train,train_preds))
```

```python
print("Test Report ==>")
print(classification_report(y_test,test_preds))
```

```python
sns.heatmap(data = confusion_matrix(y_test,test_preds),annot=True)
plt.title("Test Confusion Matrix",c='r',fontsize=15)
plt.show()
```

```python
sns.heatmap(data = confusion_matrix(y_train,train_preds),annot=True)
plt.title("Train Confusion Matrix",c='r',fontsize=15)
plt.show()
```
